# Percobaan

````python
# Pendekatan array di Python

angka = [10, 20, 30, 40, 50]
print("Array dengan list:", angka) #deklarasi array menggunakan list

import array
angka = array.array('i', [10, 20, 30, 40, 50])
print("\nArray dengan modul array:", angka) #menggunakan modul array

import numpy as np
angka = np.array([10, 20, 30, 40, 50])
print("\nArray dengan numpy:", angka) #menggunakan modul numpy

# Jenis-jenis Array

angka = np.array([1, 2, 3, 4, 5])
print("\nArray satu dimensi:", angka) #Array 1 dimensi

matriks = np.array([[1, 2], [3, 4], [5, 6]])
print("\nArray dua dimensi:", matriks) # Array 2 dimensi

array3D = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("\nArray tiga dimensi:" , array3D) # Array multidimensi

# Operasi dasar array

# Deklarasi array kosong
angka = []
# Inisialisasi array dengan elemen
angka = [10, 20, 30, 40, 50]
# Akses elemen
print("\nElemen pertama:", angka[0])
print("Elemen terakhir:", angka[-1])
# Menambahkan elemen baru
angka.append (60)
print("\nSetelah append:", angka)
# Menghapus elemen
angka.remove(30)
print("Setelah remove:", angka)
# Perubahan elemen
angka = np.delete (angka, 2) # Menghapus elemen pada indeks ke-2
print("\nSetelah penghapusan:", angka)
# Penghitungan panjang array
print("\nJumlah elemen:", len(angka))
# Iterasi array
for nilai in angka:
  print("\nNilai:", nilai)
````

# Tugas

````python
# Treasure Hunt
import colorama
from colorama import Fore, Style, init

print(Style.BRIGHT + Fore.CYAN + "\nTreasure Hunt" + Style.RESET_ALL)

size = None
while size is None:
    try:
        size = int(input(f"Masukkan ukuran peta: "))
    except ValueError:
        print(Fore.RED + "Ukuran peta harus berupa angka!" + Style.RESET_ALL)
        continue

treasure_map = ["-" for _ in range(size)]

import random
print(Style.BRIGHT + Fore.GREEN + f"Selamat datang di Treasure Hunt! Peta memiliki {size} lokasi (0 hingga {size-1}).")
treasure_index = random.randint(0, size-1)

treasure_map[treasure_index] = "X"

# Memulai permainan
def play_treasure_hunt():
    print(Style.BRIGHT + Fore.GREEN + "\nHarta karun telah disembunyikan di peta! Mulai permainan!" + Style.RESET_ALL)
    print(Style.BRIGHT + "Silahkan tebak posisi treasure (0-" + str(size-1) + ")." + Style.RESET_ALL)

    # Loop permainan
    while True:
        # Menampilkan peta saat ini
        print("\nPeta saat ini:", ["_" for _ in range(size)])

        # Meminta tebakan
        try:
            guess = int(input(Style.BRIGHT + f"Masukkan tebakanmu (0-{size-1}): "))
        except ValueError:
            print(Fore.RED + "Tebakan tidak valid!" + Style.RESET_ALL)
            continue

        # Validasi tebakan
        if guess < 0 or guess >= size:
            print(Fore.RED + "Tebakan tidak valid!" + Style.RESET_ALL)
            continue

        # Validasi tebakan benar/salah
        if treasure_map[guess] == "X":
            print(Fore.GREEN + "Selamat! Kamu menemukan harta karun!" + Style.RESET_ALL)
            break
        else:
            print(Fore.RED + "Tebakanmu salah! Coba lagi." + Style.RESET_ALL)

    print(Style.BRIGHT + Fore.CYAN + "Permainan selesai!" + Style.RESET_ALL)

# Memulai permainan
play_treasure_hunt()
````

# Penjelasan Program Game Treasure Hunt

### 1. Import dan Inisialisasi
```python
import colorama
from colorama import Fore, Style, init

print(Style.BRIGHT + Fore.CYAN + "\nTreasure Hunt" + Style.RESET_ALL)
```
- Kode ini mengimpor modul `colorama` yang digunakan untuk memberikan warna pada output teks di terminal sebagai bentuk styling kodingan.
- `Style` dan `Fore` digunakan untuk mengatur gaya dan warna teks yang ditampilkan, sebagai contohnya adalah judul "Treasure Hunt" yang berwarna cyan dan stylenya bold.

### 2. Input Ukuran Peta
```python
size = None
while size is None:
    try:
        size = int(input(f"Masukkan ukuran peta: "))
    except ValueError:
        print(Fore.RED + "Ukuran peta harus berupa angka!" + Style.RESET_ALL)
        continue
```
- Di sini, program meminta pengguna untuk memasukkan ukuran peta sebagai jumlah lokasi yang akan digunakan untuk bermain gamenya.
- Terdapat **Value Exception** berupa inputan jika bukan angka, program akan menangkap `ValueError` dan meminta pengguna untuk mencoba lagi.

### 3. Membuat Peta dan Menyembunyikan Harta Karun
```python
treasure_map = ["-" for _ in range(size)]
import random
print(Style.BRIGHT + Fore.GREEN + f"Selamat datang di Treasure Hunt! Peta memiliki {size} lokasi (0 hingga {size-1}).")
```

- Modul `random` digunakan untuk mealakukan pengacakan.
- `treasure_map` adalah daftar yang berisi tanda "-" untuk setiap lokasi di peta yang berdasarkan inputan ukuran peta sebelumnya.
- `random.randint(0, size-1)` digunakan untuk memilih secara acak indeks lokasi di mana harta karun akan disembunyikan.

### 4. Menandai Harta Karun di Peta
```python
treasure_map[treasure_index] = "X"
```
- Setelah posisi harta karun ditentukan, kode ini menandai lokasi tersebut dengan "X" di `treasure_map`.

### 5. Fungsi Permainan
```python
def play_treasure_hunt():
```

### 6. Loop Permainan
```python
while True:
    print("\nPeta saat ini:", ["_" for _ in range(size)])
```
- Di dalam loop ini, peta saat ini ditampilkan dengan tanda "_" untuk setiap lokasi, menunjukkan bahwa semua lokasi masih belum tertebak.

### 7. Meminta Tebakan
```python
try:

except ValueError:
    print(Fore.RED + "Tebakan tidak valid!" + Style.RESET_ALL)
```

- Program meminta pengguna untuk memasukkan tebakan mereka. Jika input bukan angka, program akan menangkap `ValueError` dan meminta pengguna untuk mencoba lagi.

### 8. Validasi Tebakan
```python
if guess < 0 or guess >= size:

    continue
```
- Kode ini memeriksa apakah tebakan berada dalam rentang yang valid. Jika tidak, pengguna diminta untuk mencoba lagi.

### 9. Memeriksa Kebenaran Tebakan
```python
if treasure_map[guess] == "X":
    print(Fore.GREEN + "Selamat! Kamu menemukan harta karun!" + Style.RE SET_ALL)
    break
else:
```

- Jika tebakan pengguna benar (yaitu, jika lokasi yang ditebak adalah tempat harta karun), program akan memberi tahu pengguna bahwa mereka telah menemukan harta karun dan keluar dari loop.
- Jika tebakan salah, pengguna akan diberi tahu dan diminta untuk mencoba lagi.

### 10. Menyelesaikan Permainan
```python
print(Style.BRIGHT + Fore.CYAN + "Permainan selesai!" + Style.RESET_ALL)
```

### 11. Memanggil Fungsi Permainan
