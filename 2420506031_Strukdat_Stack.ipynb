```python
class Stack:
    def __init__(self):
        """Inisialisasi stack sebagai list kosong."""
        self.stack = []

    def push(self, item):  # Push
        """Menambahkan item ke atas stack."""
        self.stack.append(item)

    def pop(self):  # Pop
        """Menghapus dan mengembalikan item teratas dari stack."""
        if self.isEmpty():
            return "Stack Kosong"
        return self.stack.pop()

    def peek(self):  # Peek
        """Mengembalikan item teratas tanpa menghapusnya."""
        if self.isEmpty():
            return "Stack Kosong"
        return self.stack[-1]

    def isEmpty(self):  # isEmpty
        """Memeriksa apakah stack kosong."""
        return len(self.stack) == 0

    def size(self):  # Size
        """Mengembalikan jumlah item dalam stack."""
        return len(self.stack)

# Penggunaan
if __name__ == "__main__":
    s = Stack()

    # Menambahkan item ke stack
    s.push(5)
    s.push(10)

    print(s.stack)  # Output: [5, 10]

    # Mengambil item dari stack
    print(s.pop())  # Output: 10

    # Melihat item teratas
    print(s.peek())  # Output: 5

    # Memeriksa apakah stack kosong
    print(s.isEmpty())  # Output: False

    # Memeriksa ukuran stack
    print(f"Ukuran stack saat ini: {s.size()}")  # Output: 1

    # Menghapus item dari stack
    s.pop()  # Menghapus 5

    # Memeriksa apakah stack kosong setelah menghapus semua item
    print(s.isEmpty())  # Output: True
    print(f"Ukuran stack setelah menghapus semua item: {s.size()}")  # Output: 0
```

```python
class TextEditor:
  def __init__(self):
    self.history = []

  def write(self, text):
    self.history.append(text)

  def undo(self):
    if self.history:
      return f"undo: meghapus '{self.history.pop()}"
    return "Tidak ada yang bisa diundo"

# Penggunaan
editor = TextEditor()
editor.write("Halo, ini teks pertama.")
editor.write("Ini teks kedua.")
editor.write("Ini teks ketiga.")
print(editor.history)

print(editor.undo())
print(editor.undo())
print(editor.history)
```

### Program *undo* pada kalkulator

```python
class Calc:
    def __init__(self):
        self.item = []

    def tambah(self, item):
        self.item.append(item)

    def hapus(self):
        if not self.kosong():
            return self.item.pop()
        return None

    def kosong(self):
        return len(self.item) == 0

    def lihat_teratas(self):
        if not self.kosong():
            return self.item[-1]
        return None

    def riwayat(self):
        return self.item


def kalkulator():
    s = Calc()

    # Meminta input NPM dari pengguna
    while True:
        npm = input("Masukkan NPM Anda (10 digit): ")
        if len(npm) == 10 and npm.isdigit():
            # Mengambil dua digit terakhir dari NPM sebagai nilai awal
            nilai_sekarang = int(npm[-2:])
            break
        else:
            print("NPM tidak valid. Pastikan NPM terdiri dari 10 digit.")

    while True:
        print(f"\nNilai saat ini: {nilai_sekarang}")
        print("Pilih operasi:")
        print("1. Tambah")
        print("2. Kurang")
        print("3. Kali")
        print("4. Bagi")
        print("5. Undo")
        print("6. Lihat Riwayat")
        print("7. Keluar")

        pilihan = input("Masukkan pilihan (1-7): ")

        if pilihan == '7':
            print("Keluar dari kalkulator")
            break

        if pilihan == '6':
            riwayat = s.riwayat()
            if riwayat:
                print("Riwayat operasi:")
                for idx, (nilai, operasi, angka) in enumerate(riwayat):
                    print(f"{idx + 1}. {nilai} {operasi} {angka}")
            else:
                print("Riwayat kosong.")
            continue

        if pilihan == '5':
            operasi_terakhir = s.hapus()
            if operasi_terakhir:
                nilai_sekarang = operasi_terakhir[0]
                print(f"Undo dilakukan. Nilai kembali ke: {nilai_sekarang}")
            else:
                print("Tidak ada operasi untuk di-undo.")
            continue

        try:
            angka = float(input("Masukkan angka: "))
        except ValueError:
            print("Input tidak valid. Silakan masukkan angka yang benar.")
            continue

        if pilihan == '1':
            s.tambah((nilai_sekarang, 'tambah', angka))
            nilai_sekarang += angka
            print(f"{nilai_sekarang} = {nilai_sekarang} + {angka}")

        elif pilihan == '2':
            s.tambah((nilai_sekarang, 'kurang', angka))
            nilai_sekarang -= angka
            print(f"{nilai_sekarang} = {nilai_sekarang} - {angka}")

        elif pilihan == '3':
            s.tambah((nilai_sekarang, 'kali', angka))
            nilai_sekarang *= angka
            print(f"{nilai_sekarang} = {nilai_sekarang} * {angka}")

        elif pilihan == '4':
            if angka != 0:
                s.tambah((nilai_sekarang, 'bagi', angka))
                nilai_sekarang /= angka
                print(f"{nilai_sekarang} = {nilai_sekarang} / {angka}")
            else:
                print("Error: Pembagian dengan nol tidak diperbolehkan.")
                continue

        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    kalkulator()
```

# Latihan 2
### Implementasi Stack dengan collections.deque - Navigasi History Browser.

```python
from collections import deque

class browserHistory:
  def __init__(self):
    self.history = deque()

  def visit(self, url):
    self.history.append(url)
    print(f"Mengunjungi:{url}")

  def back(self):
    if self.history:
      print(f"Kembali ke: {self.history.pop()}")
    else:
      print("Tidak ada riwayat!")

# Penggunaan
browser = browserHistory()
browser.visit("google.com")
browser.visit("youtube.com")
browser.visit("github.com")

browser.back()
browser.back()
```

```python
class HistoApp:
    def __init__(self):
        self.item = []

    def tambah(self, item):
        self.item.append(item)

    def hapus(self):
        if not self.kosong():
            return self.item.pop()
        return None

    def kosong(self):
        return len(self.item) == 0

def navigasi_aplikasi():
    history = HistoApp()

    # Meminta jumlah fitur yang akan dibuat
    while True:
        try:
            jumlah_fitur = int(input("Berapa banyak fitur yang akan dibuat? "))
            if jumlah_fitur > 0:
                break
            else:
                print("Jumlah fitur harus lebih dari 0.")
        except ValueError:
            print("Input tidak valid. Silakan masukkan angka.")

    # Loop untuk meminta input bulan lahir dan NPM untuk setiap fitur
    fitur_aplikasi = []
    for i in range(jumlah_fitur):
        print(f"\nInput untuk fitur {i + 1}:")

        # Meminta input bulan lahir dari pengguna
        bulan_lahir = input("Masukkan bulan lahir Anda (misal: Januari, Februari, dst.): ").strip().lower()

        # Meminta input NPM dari pengguna
        while True:
            npm = input("Masukkan NPM Anda (10 digit): ")
            if len(npm) == 10 and npm.isdigit():
                break
            else:
                print("NPM tidak valid. Pastikan NPM terdiri dari 10 digit.")

        # Mengambil huruf pertama dari bulan lahir dan dua digit terakhir dari NPM
        huruf_bulan = bulan_lahir[0] if bulan_lahir else 'X'  # Default ke 'X' jika input kosong
        dua_digit_npm = npm[-2:]

        # Membuat nama fitur aplikasi
        fitur_aplikasi.append(f"{huruf_bulan}{dua_digit_npm}")

    # Simulasi mengunjungi fitur aplikasi
    while True:
        print("\nFitur aplikasi yang tersedia:")
        for i, fitur in enumerate(fitur_aplikasi, start=1):
            print(f"{i}. {fitur}")

        pilihan = input("\nPilih fitur aplikasi (1-{}) atau ketik 'back' untuk kembali atau 'exit' untuk keluar: ".format(jumlah_fitur))

        if pilihan.lower() == 'exit':
            print("Keluar dari aplikasi.")
            break

        if pilihan.lower() == 'back':
            fitur_terakhir = history.hapus()
            if fitur_terakhir:
                print(f"Kembali ke fitur: {fitur_terakhir}")
            else:
                print("Tidak ada history.")
            continue

        try:
            pilihan = int(pilihan)
            if 1 <= pilihan <= len(fitur_aplikasi):
                fitur_dikunjungi = fitur_aplikasi[pilihan - 1]
                history.tambah(fitur_dikunjungi)
                print(f"Anda mengunjungi fitur: {fitur_dikunjungi}")
            else:
                print("Pilihan tidak valid. Silakan coba lagi.")
        except ValueError:
            print("Input tidak valid. Silakan masukkan angka atau 'back'.")

if __name__ == "__main__":
    navigasi_aplikasi()
```

### Implementasi Stack dengan queue. LifoQueue - Sistem Pencadangan File

```python
from queue import LifoQueue

class BackupSystem:
  def __init__(self):
    self.backups = LifoQueue ()
  def backup(self, version):
    self.backups.put(version)

  def latest_backup (self):
    if not self.backups.empty():
      return f"Versi terbaru: {self.backups.queue[-1] }"
    return "Tidak ada cadangan!"

# Contoh Penggunaan
backup_system = BackupSystem()
backup_system.backup("Backup_1")
backup_system.backup ("Backup_2")
backup_system.backup("Backup_3")

print(backup_system.latest_backup())
```

### Sistem Penyimpanan Draft Email

```python
class EmailDraft:
    def __init__(self):
        self.drafts = []

    def simpan_draft(self, nama_mahasiswa, npm, isi_email):
        subjek = self.buat_subjek(nama_mahasiswa, npm)
        draft = {
            'subjek': subjek,
            'isi': isi_email
        }
        self.drafts.append(draft)
        print(f"Draft email disimpan dengan subjek: {subjek}")

    def hapus_draft(self):
        if not self.kosong():
            return self.drafts.pop()
        return None

    def kosong(self):
        return len(self.drafts) == 0

    def lihat_draft_terakhir(self):
        if not self.kosong():
            return self.drafts[-1]
        return None

    def lihat_history(self):
        if self.kosong():
            print("Tidak ada draft yang tersedia.")
        else:
            print("\nRiwayat Draft:")
            for idx, draft in enumerate(self.drafts, start=1):
                print(f"{idx}. Subjek: {draft['subjek']}, Isi: {draft['isi']}")

    def buat_subjek(self, nama_mahasiswa, npm):
        dua_huruf_nama = nama_mahasiswa[:2].upper()
        dua_digit_npm = npm[-2:]
        return f"{dua_huruf_nama}{dua_digit_npm}"

def main():
    email_draft = EmailDraft()

    while True:
        print("\nMenu:")
        print("1. Buat email dan simpan dalam draft")
        print("2. Lihat riwayat draft")
        print("3. Hapus draft terakhir")
        print("4. Keluar")

        pilihan = input("Pilih opsi (1-4): ")

        if pilihan == '1':
            # Meminta input nama mahasiswa dan NPM dari pengguna
            nama_mahasiswa = input("Masukkan nama mahasiswa: ").strip()

            while True:
                npm = input("Masukkan NPM Anda (10 digit): ")
                if len(npm) == 10 and npm.isdigit():
                    break
                else:
                    print("NPM tidak valid. Pastikan NPM terdiri dari 10 digit.")

            # Meminta isi email
            isi_email = input("Masukkan isi email: ")
            email_draft.simpan_draft(nama_mahasiswa, npm, isi_email)

        elif pilihan == '2':
            email_draft.lihat_history()

        elif pilihan == '3':
            if email_draft.kosong():
                print("Tidak ada draft yang dapat dihapus.")
            else:
                email_draft.hapus_draft()
                print("Draft terakhir telah dihapus.")

        elif pilihan == '4':
            print("Keluar dari aplikasi.")
            break

        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```

### Sistem Antrean Tiket Bioskop

```python
class CustomerService:
    def __init__(self):
        self.tickets = []

    def add_ticket(self, ticket):
        self.tickets.append(ticket)
        print(f"Tiket '{ticket}' telah ditambahkan ke antrean.")

    def process_ticket(self):
        if not self.is_empty():
            return f"Memproses tiket: {self.tickets.pop()}"
        return "Tidak ada tiket yang harus diproses."

    def is_empty(self):
        return len(self.tickets) == 0

    def lihat_antrean(self):
        if self.is_empty():
            return "Antrean kosong."
        return f"Antrean saat ini: {self.tickets}"

def buat_nama_film(bulan_lahir, npm):
    # Mengambil huruf pertama dari bulan lahir
    huruf_bulan = bulan_lahir[0].upper()
    # Mengambil dua digit terakhir dari NPM
    dua_digit_npm = npm[-2:]
    return f"{huruf_bulan}{dua_digit_npm}"

def main():
    service = CustomerService()

    # Membeli minimal 3 tiket
    service.add_ticket(buat_nama_film("Januari", "12345678"))
    service.add_ticket(buat_nama_film("Februari", "87654321"))
    service.add_ticket(buat_nama_film("Maret", "11223344"))

    # Melihat antrean saat ini
    print(service.lihat_antrean())

    # Memproses tiket
    print(service.process_ticket())
    print(service.process_ticket())

    # Melihat antrean setelah memproses beberapa tiket
    print(service.lihat_antrean())

    # Memproses tiket hingga antrean kosong
    print(service.process_ticket())
    print(service.process_ticket())  # Coba memproses tiket lagi

if __name__ == "__main__":
    main()
```

# Tugas
### Simulasi Tumpukan Barang dalam Gudang

```python
class Gudang:
    def __init__(self, kapasitas):
        self.kapasitas = kapasitas
        self.barang = []

    def push(self, item):
        """Menambahkan barang ke gudang jika belum penuh."""
        if len(self.barang) < self.kapasitas:
            self.barang.append(item)
            print(f"Barang '{item}' telah ditambahkan ke gudang.")
        else:
            print("Gudang sudah penuh! Tidak bisa menambahkan barang.")

    def pop(self):
        """Mengambil barang terakhir yang ditambahkan ke gudang."""
        if not self.is_empty():
            return self.barang.pop()
        return "Gudang kosong! Tidak ada barang untuk diambil."

    def is_empty(self):
        """Memeriksa apakah gudang kosong."""
        return len(self.barang) == 0

    def lihat_gudang(self):
        """Mengembalikan daftar barang yang ada di gudang."""
        if self.is_empty():
            return "Gudang kosong."
        return f"Barang dalam gudang: {self.barang}"

def main():
    while True:
        # Meminta pengguna untuk memasukkan NPM
        print("Simulasi Tumpukan Barang dalam Gudang")
        npm = input("Masukkan NPM Anda (8 digit): ")

        # Memastikan NPM memiliki panjang yang benar
        if len(npm) == 8 and npm.isdigit():
            # Mengambil dua digit terakhir dari NPM sebagai kapasitas gudang
            kapasitas_gudang = int(npm[-2:])
            break  # Keluar dari loop jika NPM valid
        else:
            print("NPM tidak valid. Pastikan NPM terdiri dari 8 digit. Silakan coba lagi.")

    gudang = Gudang(kapasitas_gudang)

    while True:
        print("\n=== Simulasi Tumpukan Barang dalam Gudang ===")
        print("1. Tambah Barang")
        print("2. Ambil Barang")
        print("3. Lihat Barang di Gudang")
        print("4. Keluar")

        pilihan = input("Pilih opsi (1-4): ")

        if pilihan == '1':
            barang = input("Masukkan nama barang yang ingin ditambahkan: ")
            gudang.push(barang)

        elif pilihan == '2':
            barang_diambil = gudang.pop()
            print(barang_diambil)

        elif pilihan == '3':
            print(gudang.lihat_gudang())

        elif pilihan == '4':
            print("Keluar dari permainan.")
            break

        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
