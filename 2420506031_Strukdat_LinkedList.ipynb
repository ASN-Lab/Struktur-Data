```python
# Latihan 2
print("\nLatihan 2")

# Definisi Node dalam Doubly Linked List
class PageNode:
    def __init__(self, url):
        self.url = url # Nama halaman
        self.next = None        # Pointer ke node selanjutnya
        self.prev = None        # Pointer ke node sebelumnya

# Definisi Doubly Linked List untuk riwayat browsing
class BrowserHistory:
    def __init__(self):
        self.current = None # Node saat ini dalam linked list

    # Menambahkan halaman baru ke linked list
    def visit_page(self, url):
        new_page = PageNode(url)
        if not self.current:
            self.current = None # Menjadikan halaman baru sebagai node pertama
        else:
            new_page.prev = self.current
            self.current.next = new_page
            self.current = new_page # Pindah ke halaman baru

    # Menampilkan riwayat browsing
    def show_history(self):
        if not self.current:
            print("Riwayat browsing kosong.")
            return

        print("Riwayat Browsing:")
        temp = self.current
        while temp.prev: # Pindah ke halaman pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.url}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke halaman berikutnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"Kembali ke halaman: {self.current.url}")
        else:
            print("Tidak ada halaman sebelumnya.")

    # Navigasi maju ke halaman sebelumnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"Maju ke halaman: {self.current.url}")
        else:
            print("Tidak ada halaman berikutnya.")
# Main #

# Membuat objek riwayat peramban
history = BrowserHistory()

# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("Masukkan jumlah halaman yang dikunjungi: "))

# Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page (url)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back ()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\nKeluar dari riwayat browser.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```

```python
# Latihan 3
print("\nLatihan 3")

# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data # Menyimpan nilai node
        self.next = None # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data): # Indented to be part of LinkedList class
        if position < 0:
            print(" Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print(" Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    #Menghapus elemen di awal
    def delete_at_beginning(self): # Indented to be part of LinkedList class
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self): # Indented to be part of LinkedList class
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position): # Indented to be part of LinkedList class
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range (position - 1):
            if not temp or not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
            temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key) :  # Indented to be part of LinkedList class
        temp = self.head
        while temp:
            if temp.data == key:
                print(f" Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print (f" Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse (self):  # Indented to be part of LinkedList class
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display (self):  # Indented to be part of LinkedList class
        if not self.head:
            print(" Daftar kosong!")
            return
        temp = self.head
        while temp:
            print (temp.data, end=" -> ")
            temp = temp.next
        print("None")

# Program Utama (Main)

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range (num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n Pilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning ()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int (input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```

```python
# Definisi Node dalam Circular Linked List
class GuardNode:
    def __init__(self, name):
        self.name = name  # Nama satpam
        self.next = None  # Pointer ke node berikutnya

# Definisi Circular Linked List untuk jadwal piket
class GuardSchedule:
    def __init__(self):
        self.head = None  # Node pertama dalam linked list

    # Menambahkan satpam ke dalam linked list
    def add_guard(self, name):
        new_guard = GuardNode(name)
        if not self.head:
            self.head = new_guard
            new_guard.next = self.head  # Menghubungkan ke dirinya sendiri
        else:
            current = self.head
            while current.next != self.head:
                current = current.next  # Menemukan node terakhir
            current.next = new_guard
            new_guard.next = self.head  # Menghubungkan node baru ke head

    # Mencetak jadwal piket selama 14 hari
    def print_schedule(self):
        if not self.head:
            print("Tidak ada satpam yang terdaftar.")
            return

        current = self.head
        print("Jadwal Piket Satpam selama 14 hari:")
        for day in range(1, 15):  # 14 hari
            print(f"Hari {day}: {current.name}")
            current = current.next  # Berpindah ke satpam berikutnya

# Main Program
if __name__ == "__main__":
    # Membuat objek jadwal piket
    schedule = GuardSchedule()

    # Meminta pengguna memasukkan nama satpam
    for i in range(4):
        name = input(f"Masukkan nama satpam ke-{i + 1}: ")
        schedule.add_guard(name)

    # Mencetak jadwal piket
    schedule.print_schedule()
```
